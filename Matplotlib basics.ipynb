```python
plt.subplot(2,1,1) #2 rows, 1 column, subplot number 1
plt.plot([1,2,3,4],[5,6,7,8])
plt.title("1st Subplot")

plt.subplot(2,1,2)
plt.plot(x,y)
plt.title("2nd Subplot")

plt.suptitle("Slight difference")
plt.subplots_adjust(hspace=0.5)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6)) #SubplotS for multiple plots in a figure
ax[0,1].plot([1,2,3,4],[5,6,7,8], 'r^')
ax[0,1].set_title("Squares")
ax[1,0].plot(x,y,'go')
ax[1,0].set_title("Cubes")

plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.suptitle("Mo' plots, mo' problems")
plt.show()
```

### Different types of graphs

##### 1. Bar graphs - Used to show data associated with the categorical variables

```python
divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"]
division_average_marks = [70, 82, 73, 65, 68]

plt.bar(divisions, division_average_marks, color="purple")
plt.title("Bar Graph")
plt.xlabel("Division")
plt.ylabel("Marks")

plt.show()
```

```python
variance = [5,8,7,6,4] #I'm not really sure what a variance is in statistics. Will find out

plt.barh(divisions, division_average_marks, xerr=variance, color="brown") #Horizontal
plt.title("Horizontal Bar")
plt.xlabel("Marks")
plt.ylabel("Division")
plt.show()
```

```python
boys_average_marks = [68,67,77,61,70]

index = np.arange(5)
width = 0.30

plt.bar(index, division_average_marks, width, color="pink", label="Division Marks")
plt.bar(index+width, boys_average_marks, width, color="brown", label="Boys Marks") #1st argument: So that the bars 
                                                                                    #are next to each other
plt.title("Horizontally Stacked Bars")

plt.ylabel("Marks")
plt.xlabel("Divisions")
plt.xticks(index+width/2, divisions) #Positions of xlabels

plt.legend(loc="best")#Labels for the bars

plt.show()
```

```python
divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"]
girls_average_marks = [72, 97, 69, 69, 66]
boys_average_marks = [68,67,77,61,70]

index = np.arange(5)
width = 0.30
plt.bar(index, girls_average_marks, width, color="pink", label="Girls Marks",bottom=boys_average_marks)#Assign bottom value
plt.bar(index, boys_average_marks, width, color="brown", label="Boys Marks") #1st argument: So that the bars 
                                                                             #are stacked onto each other
plt.title("Vertically Stacked Bars")

plt.ylabel("Marks")
plt.xlabel("Divisions")
plt.xticks(index, divisions) #Positions of xlabels

plt.legend(loc="best")#Labels for the bars

plt.show()
```

##### 2. Pie charts 

```python
firms = ["Firm A", "Firm B", "Firm C", "Firm D", "Firm E"]
market_share = [20,25,15,10,20]

Explode = [0,0.1,0,0,0]

plt.pie(market_share, explode=Explode,labels=firms, shadow=True, startangle=45)
plt.axis('equal')
plt.title("Pie Chart")
plt.legend(title="List of firms", loc="best")

plt.show()
```

```python
firms = ["Firm A", "Firm B", "Firm C", "Firm D", "Firm E"]
market_share = [20,25,15,10,20]
colors = ['b', 'g', 'r', 'c', 'y']

Explode = [0,0.1,0,0,0]

plt.pie(market_share, colors=colors,explode=Explode,labels=firms, shadow=True, startangle=90, 
        autopct="%1.1f%%", counterclock=False) #put percentage
plt.axis('equal')
plt.title("Pie Chart")
plt.legend(title="List of firms", loc="upper right", bbox_to_anchor=(1.1,1)) #upper left, lower right/left, location of legend

plt.show()
```

#### 4. Scatter Plots and 3D plotting - come in handy in visualizing a problem of regression.

```python
height = np.array([167, 170, 149, 165, 155, 180, 166, 146,
                  159, 185, 145, 168, 172, 181, 169])
weight = np.array([86, 74, 66, 78, 68, 79, 90, 73,
                  70, 88, 66, 84, 67, 84, 77])

plt.xlim(149,200)
plt.ylim(60,100)
plt.scatter(height, weight, color="darkgreen")
plt.title("Scatter Plot")
plt.xlabel("Height")
plt.ylabel("Weight")

plt.show()
```

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')
ax.scatter3D(height,weight, color="purple")
ax.set_xlabel("Height")
ax.set_ylabel("Weight")

plt.show()
```

#### 5. Heat map - Usually the darker shades of the chart represent higher values than the lighter shade.


```python
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest)

# We want to show all ticks...
ax.set_xticks(np.arange(len(farmers)))
ax.set_yticks(np.arange(len(vegetables)))
# ... and label them with the respective list entries
ax.set_xticklabels(farmers)
ax.set_yticklabels(vegetables)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")
        
ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()


plt.show()
```

#### 6. Bubble chart - 
A bubble plot is a scatterplot where a third dimension is added: the value of an additional variable is represented through the size of the dots. You need 3 numerical variables as input: one is represented by the X axis, one
by the Y axis, and one by the size. Do not forget to provide a legend to make possible the link between the size and the value. Note that too many bubble make the chart hard to read, so this type of representation is usually
not recommended for big amount of data. Last but not least, note that the area of the circles must be proportional to the value, not to the radius, to avoid exaggerate the variation in your data.

```python
import matplotlib.pyplot as plt
import numpy as np
 
# create data
x = np.random.rand(40)
y = np.random.rand(40)
z = np.random.rand(40)
colors = np.random.rand(40) 
# use the scatter function
plt.scatter(x, y, s=z*1000,c=colors)
plt.show()
 
```
